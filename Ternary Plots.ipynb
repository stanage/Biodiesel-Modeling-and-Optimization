```python
import pandas as pd
import statsmodels.api as sm
from itertools import combinations
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

## Density Ternary Contour Plot

```python
import plotly.figure_factory as ff
import numpy as np

data_oil = pd.read_excel("Oil and catalyst Mixture designs plot.xlsx", sheet_name="Oil design")

test_data = data_oil.values

a = test_data[:, 0]
b = test_data[:, 1]
c = test_data[:, 2]

v = test_data[:, 3]

fig = ff.create_ternary_contour(
    np.array([a, b, c]), v,
    pole_labels=['A', 'B', 'C'],
    ncontours=5,
    colorscale='Greys',
    showscale=False,
    coloring="lines",
    showmarkers=True
    
)

fig.update_layout(
    ternary=dict(
        sum=100,
        aaxis=dict(title='WVO', ticks='inside', tickfont=dict(size=10)),
        baxis=dict(title='NSO', ticks='inside', tickfont=dict(size=10)),
        caxis=dict(title='CSO', ticks='inside', tickfont=dict(size=10)),
    ),
)

fig.show()

```
